## Sampling from a Biased Population

```python
# Import the packages that we will be using for the tutorial
import numpy as np # for sampling for the distributions
import matplotlib.pyplot as plt # for basic plotting 
import seaborn as sns; sns.set() # for plotting of the histograms
import warnings
warnings.filterwarnings("ignore")

mean_uofm = 155 # mean of weight of students who do not go to gym
sd_uofm = 5
mean_gym = 185 # mean of weight of students who go to gym
sd_gym = 5 
gymperc = .3 # percentage of students go to gym
totalPopSize = 40000

# Create the two subgroups using normal distributions
uofm_students = np.random.normal(mean_uofm, sd_uofm, int(totalPopSize * (1 - gymperc))) # gym not goer
students_at_gym = np.random.normal(mean_gym, sd_gym, int(totalPopSize * (gymperc))) # gym goer

# Create the population from the subgroups
population = np.append(uofm_students, students_at_gym)

# Set up the figure for plotting
plt.figure(figsize=(10,12))

# Plot the UofM students only
plt.subplot(3,1,1)
sns.distplot(uofm_students)
plt.title("UofM Students Only")
plt.xlim([140,200])

# Plot the Gym Goers only
plt.subplot(3,1,2)
sns.distplot(students_at_gym)
plt.title("Gym Goers Only")
plt.xlim([140,200])

# Plot both groups together
plt.subplot(3,1,3)
sns.distplot(population)
plt.title("Full Population of UofM Students")
plt.axvline(x = np.mean(population)) #draws a vertical line
plt.xlim([140,200])

plt.show()
```

## What Happens if We Sample from the Entire Population?
We will sample randomly from all students at the University of Michigan.

```python
from past.builtins import xrange

# Simulation parameters
numberSamps = 5000 # number of means
sampSize = 50 # number of sample from overall population

# Get the sampling distribution of the mean from only the gym
mean_distribution = np.empty(numberSamps)
for i in xrange(numberSamps):
    random_students = np.random.choice(population, sampSize) # random selection from entire population
    mean_distribution[i] = np.mean(random_students)
    
# Plot the population and the biased sampling distribution
plt.figure(figsize = (10,8))

# Plotting the population again
plt.subplot(2,1,1)
sns.distplot(population)
plt.title("Full Population of UofM Students")
plt.axvline(x = np.mean(population))
plt.xlim([140,200])

# Plotting the sampling distribution
plt.subplot(2,1,2)
sns.distplot(mean_distribution)
plt.title("Sampling Distribution of the Mean Weight of All UofM Students")
plt.axvline(x = np.mean(population))
plt.axvline(x = np.mean(mean_distribution), color = "black")
plt.xlim([140,200])

plt.show()
```

What we end up doing though, is we end up plotting the sample means as well. So, what this is, this is 5,000 of those sampling distributions, 5,000 those of mean distributions that we plotted from above, and then we can see that if I take 50 people and I plot the mean, let's say that I get this value right here, this bar would go up by one, and then I take another 50 people, I end up saying lets say now the mean is, let's say 161, and I'll put another piece here. If I do that 5,000 times, I'd see this histogram. We can see that the black line, the mean of my sampling distribution falls almost exactly on top of my blue line, the mean of the population. This is to say that if I were to repeat this process of just sampling 50 random students, calculate the mean, I should expect on average to get a mean for my sample that's almost exactly equal to my population if I sample truly from everyone in my population, if I truly take samples from every one of my population. So, this is what we expect. This is why sampling is good whenever we take probability-based sample, because we get means that are roughly equivalent to what we hope they are actually trying to measure. The question though, now becomes what happens if we take a non representative sample?

## What Happens if We take a Non-Representative Sample?
What happens if I only go to the gym to get the weight of individuals, and I don't sample randomly from all students at the University of Michigan?

```python
# Simulation parameters
numberSamps = 5000
sampSize = 3

# Get the sampling distribution of the mean from only the gym
mean_distribution = np.empty(numberSamps)
for i in xrange(numberSamps):
    random_students = np.random.choice(students_at_gym, sampSize)
    mean_distribution[i] = np.mean(random_students) 
    

# Plot the population and the biased sampling distribution
plt.figure(figsize = (10,8))

# Plotting the population again
plt.subplot(2,1,1)
sns.distplot(population)
plt.title("Full Population of UofM Students")
plt.axvline(x = np.mean(population))
plt.xlim([140,200])

# Plotting the sampling distribution
plt.subplot(2,1,2)
sns.distplot(mean_distribution)
plt.title("Sampling Distribution of the Mean Weight of Gym Goers")
plt.axvline(x = np.mean(population))
plt.axvline(x = np.mean(students_at_gym), color = "black")
plt.xlim([140,200])

plt.show()
```

We notice that I'm only pulling from people from this graph right here, and we see that my sampling distribution has a very biased mean. So, we can see that the blue line what the mean should be is very far away from the black line, which is what my probability samples, my non-probability samples. In this case, I used convenience sample. I am only going to measure people at the gym, and as a result, I got an incredibly biased estimate of the full population because it turns out that 70 percent of the people are people that I would've never been able to sample at the gym or outside of the gym. So, this is some of the dangers of using things like convenience samples. Sometimes, you're not taking samples of the entire population but instead of only select subsets, and as a result we can get really biased estimates.
